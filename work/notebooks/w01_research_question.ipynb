## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Answers:
*   The Decision: Which specific pages should the content team review and update first? (Unit of analysis: a single content page).
*   The Action: A human editor reviews the top-ranked pages in the queue, reads the provided "reason codes" (e.g., high impressions but dropping traffic), and decides whether to update the page's content, metadata, or leave it alone.
*  The Cost of a Wrong Call: Wasted editor hours. If the model prioritizes pages that are only experiencing temporary seasonal dips or low-volume noise, the editor spends valuable time fixing something that wasn't broken, while truly decaying pages are left untouched.
*  Why ML Helps: A simple rule (like trend_direction == 'down') captures too much noise and surfaces too many pages for a human to realistically review. ML can weigh multiple signals simultaneously (age, search volume, CTR, engagement) to separate true decay from normal variance and rank them strictly by priority.



```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

down_pages = len(df[df['trend_direction'] == 'down'])
total_pages = len(df)

print(f"Pages with trend_direction == 'down': {down_pages:,} out of {total_pages:,}")
print("Conclusion: A simple 'down' rule flags too many pages for a human editor to review. We need ML to rank and prioritize the most important ones.")
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Answers:
To prove this lane is worth 7 weeks of work, we need to show that finding the right page to refresh is a "needle in a haystack" problem.
If we look at the data, a huge portion of the inventory has decent visibility, but only a fraction of those are actively losing traction (declining). A human editor cannot manually monitor thousands of high-visibility pages every day. ML is perfectly suited to filter this massive pool of active pages down to a ranked list of the most urgent decay risks, saving hundreds of hours of manual auditing.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/Rmdsketch/flyrank-ml-internship/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# Number 1: How many pages have meaningful visibility? (e.g., >= 500 impressions in 90 days)
high_vis_pages = len(df[df['impressions_90d'] >= 500])

# Number 2: Out of those highly visible pages, how many are actually declining?
declining_high_vis = len(df[(df['impressions_90d'] >= 500) & (df['trend_direction'] == 'down')])

# Number 3: Calculate the percentage of high-visibility pages that are dropping
percentage = (declining_high_vis / high_vis_pages) * 100 if high_vis_pages > 0 else 0

print(f"1. Meaningful visibility pages (>= 500 impressions): {high_vis_pages:,}")
print(f"2. High-visibility pages that are actively declining: {declining_high_vis:,}")
print(f"3. The needle in the haystack: Only {percentage:.1f}% of visible pages need urgent decay review.")
print("\nConclusion: Finding these specific pages manually across a 30,000+ page inventory is impossible. A ranked queue solves this.")
```
